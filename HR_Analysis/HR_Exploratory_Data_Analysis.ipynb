### Grouping Age of employees with respect to some important aspects with age:

```python
age=pd.DataFrame(data.groupby("Age")[["MonthlyIncome","Education","JobLevel","JobInvolvement","PerformanceRating","JobSatisfaction","EnvironmentSatisfaction","RelationshipSatisfaction","WorkLifeBalance","DailyRate","MonthlyRate"]].mean())
```

```python
age
```

### Grouping with respect to gender to seek some rwelevant information:

```python
gender=pd.DataFrame(data.groupby("Gender")[["MonthlyIncome","Education","JobLevel","JobInvolvement","PerformanceRating","JobSatisfaction","EnvironmentSatisfaction","RelationshipSatisfaction","WorkLifeBalance","DailyRate","MonthlyRate"]].mean())
```

```python
gender
```

### Salary with respect to Job Role

```python
income=pd.DataFrame(data.groupby('JobRole').MonthlyIncome.mean().sort_values(ascending=True))
```

```python
income
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=income.index,y=income.MonthlyIncome)
plt.xticks(rotation=60)
plt.xlabel('job Role')
plt.ylabel('Monthly Salry')
plt.show()
```

### Percent salary hike with respect to Job Role

```python
p_hike=pd.DataFrame(data.groupby('JobRole').PercentSalaryHike.mean().sort_values(ascending=True))
```

```python
p_hike
```

```python
sns.barplot(x=p_hike.index,y=p_hike.PercentSalaryHike)
plt.xticks(rotation=90);
```
